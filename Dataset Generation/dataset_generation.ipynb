## Generation of Simple Sentences

```python
objects = ['circle','triangle','square','rectangle','diamond', ' star', 'pentagon', 'hexagon', 'octagon', 'cube']
colours = ['red', 'blue', 'green', 'yellow', 'orange', 'purple', 'pink']
size = ['big', 'medium', 'small']
relations = ['above', 'below', 'left of', 'right of']
sentence_templates = [
    'a {bundle1} is {relation} a {bundle2}',
    'the {bundle1} is {relation} a {bundle2}',
    'a {bundle1} can be seen {relation} a {bundle2}',
    '{relation} a {bundle1} is a {bundle2}',
    'a {bundle1} can be seen {relation} a {bundle2}'
]

def creator(objects:[List], colours:[list], size:[list], sentence_templates:[list]) -> [list]:
    """
    Creates list of structured sentences where each bundle represents a concatenated str of <size> <colour> <object> 
    Output is list of structured sentences
    """
    sentences = []
    for template in sentence_templates:
        for object1 in objects:
            for colour1 in colours:
                for s1 in size:
                    for object2 in objects:
                        for colour2 in colours:
                            for s2 in size:
                                for rel in relations:
                                    
                                    bundle1 = f'{s1} {colour1} {object1}'
                                    relation = f'{rel}'
                                    bundle2 = f'{s2} {colour2} {object2}'
                                    if bundle1 == bundle2:
                                        continue
                                    sentences.append(template.format(bundle1=bundle1, relation=relation, bundle2=bundle2))
    return sentences


sentences = creator(objects, colours, size, sentence_templates)
print(len(sentences))
print(sentences[:5])
```

```python
def gen_numbers(colours:[list], size:[list]) ->[list]:
    sentences = []
    for colour in colours:
        for s in size:
            for num in range(100):
                sentence = f'{s} {colour} {num}'
                sentences.append(sentence)
    return sentences

sentences = gen_numbers(colours, size)
print(sentences[200:206])
```

```python
import random
class TicTacToe:
    def __init__(self):
        self.board = [  '-', '-', '-',
                        '-', '-', '-',
                        '-', '-', '-']
        self.player_token = 'X'
        self.dataset = []

    def print_board(self):
        print(f'{self.board[0]} | {self.board[1]} | {self.board[2]}')
        print('---------')
        print(f'{self.board[3]} | {self.board[4]} | {self.board[5]}')
        print('---------')
        print(f'{self.board[6]} | {self.board[7]} | {self.board[8]}')
    
    def check_winner(self):
        winner = [
            (0,1,2), (3,4,5), (6,7,8), 
            (0,3,6), (1,4,7), (2,5,8),
            (0,4,8), (6,4,2)
        ]
        for v1, v2, v3 in winner:
            if self.board[v1] == self.board[v2] == self.board[v3] and self.board[v1] !='-':
                return True
            
        if '-' not in self.board:
            return True
        return False
        
    def make_move(self):
        empty = [idx for idx, val in enumerate(self.board) if val == '-']
        if len(empty) == 0:
            return False
        num = random.choice(empty)
        self.board[num] = self.player_token
        return True

    def switch_player(self):
        if self.player_token == 'X':
            self.player_token = 'O'
        else:
            self.player_token = 'X'

    def add_game_state(self):
        self.dataset.append(self.board.copy())
        
    def play_game(self):
        while True:
            self.make_move()
            print('_____________________')
            self.print_board()
            
            self.add_game_state()
            if self.check_winner() == True:
                    break
            self.switch_player()

```

```python
tictactoe = TicTacToe()
tictactoe.play_game()
    
```
